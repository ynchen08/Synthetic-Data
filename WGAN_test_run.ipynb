### Create a function for implementing the WGAN algorithm on the sample real dataset

```python
def PMC_wgan(start,end,seed,batch_prop=0.3,
             optimizer=torch.optim.Adam,
             critic_d_hidden=[128,512,512,128],critic_dropout=0,
             critic_steps=15, critic_lr=0.0001,critic_gp_factor=5,
             generator_d_hidden=[128,512,512,128],generator_dropout=0.1,
             generator_lr=0.0001,generator_d_noise='generator_d_output',
             generator_optimizer='optimizer', max_epochs=1000, 
             test_set_size=16):
    #Set up data wrapper
    PMC_names=list(originaldata)[1:53]
    week_index=np.arange(52)
    want_weeks=np.arange(start-1,end)
    exclude_weeks=np.setdiff1d(week_index,want_weeks)
    exclude_PMC=list(np.array(PMC_names)[exclude_weeks])
    inputdata=originaldata.drop(exclude_PMC,axis=1)

    cont_vars=list(inputdata)
    cont_vars.remove('ID')
    cont_vars.remove('Class')

    cat_vars=[]
    cont_LB={d: 0 for d in cont_vars}
    cont_UB={d: 1 for d in cont_vars}
    context_var=['Class']

    data_wrapper=wgan.DataWrapper(inputdata, 
                                      continuous_vars=cont_vars, categorical_vars=cat_vars,
                                      context_vars=context_var,
                                      continuous_lower_bounds=cont_LB,
                                      continuous_upper_bounds=cont_UB)
    #Preprocess input data
    x, context =data_wrapper.preprocess(inputdata)
    
    #Set up the specs for WGAN model
    torch.manual_seed(seed) #note sure if these seeding commands help reproducibility
    random.seed(seed)
    np.random.seed(seed)
    specs = wgan.Specifications(data_wrapper, 
                                batch_size=round(len(inputdata)*batch_prop), 
                                optimizer=optimizer,
                                critic_d_hidden=critic_d_hidden,
                                critic_dropout=critic_dropout,
                                critic_steps=critic_steps, 
                                critic_lr=critic_lr,
                                critic_gp_factor=critic_gp_factor,
                                generator_d_hidden=generator_d_hidden,
                                generator_dropout=generator_dropout,
                                generator_lr=generator_lr,
                                generator_d_noise=generator_d_noise,
                                generator_optimizer=generator_optimizer, 
                                max_epochs=max_epochs, 
                                test_set_size=test_set_size
                               )
    
    #Set up generator and critic objects
    generator = wgan.Generator(specs)
    critic = wgan.Critic(specs)
    
    #Train the WGAN model
    start_time = time.time()
    wgan.train(generator, critic, x, context, specs)
    print("---Total run time: %s minutes ---" % ((time.time() - start_time)/60))
    #Generate synthetic dataset (n=1e4)
    sdataset = data_wrapper.apply_generator(generator, inputdata.sample(int(1e4), replace=True))
    return [sdataset,data_wrapper,generator,critic]
    
```

### Synthesize a portion of the real time series data (week 10~25) 

```python
Syn_1to25=PMC_wgan(start=1,end=25, seed=0, batch_prop=0.3,
              max_epochs=3000,
              critic_d_hidden=[128,512,512,128],
              generator_d_hidden=[128,512,512,128],
              generator_lr=0.00015,critic_lr=0.00015,
              optimizer=torch.optim.RMSprop)
```

```python
#Create a plotting function
def plot_PMC(sdataset,start,end):
    #Pre-process the real and fake data for plotting
    cont_vars=list(sdataset)
    cont_vars.remove('ID')
    cont_vars.remove('Class')
    Dat=[]
    Dat2=[]
    for i in range(max(originaldata["Class"])):
        d1=originaldata[originaldata["Class"]==i+1].loc[:,cont_vars]
        Dat.append(d1)
        d2=sdataset[sdataset["Class"]==i+1].loc[:,cont_vars]
        Dat2.append(d2.sample(d1.shape[0]))

    #Plot the real and fake data by class
    alph=0.007
    color=['blue','red','green']
    week=np.arange(start,end+1)
    
    fig,a= plt.subplots(1,2,figsize=(15,5))
    fig.tight_layout(pad=2.0)

    a[0].set_title('Trajectories by class, real data', fontsize=15)
    a[0].set_xlabel('Week')
    a[0].set_ylabel('PMC')

    for j in range(len(Dat)):
        for i in range(Dat[j].shape[0]):
            a[0].plot(week,Dat[j].iloc[i,:],color[j], linewidth=1, alpha=alph)


    a[1].set_title('Trajectories by class, synthetic data', fontsize=15)
    a[1].set_ylabel('PMC')
    a[1].set_xlabel('Week')
    for j in range(len(Dat2)):
        for i in range(Dat2[j].shape[0]):
            a[1].plot(week,Dat2[j].iloc[i,:],color[j], linewidth=1, alpha=alph)

    plt.show()
```

```python
plot_PMC(Syn_1to25[0],start=1, end=25)
```
